# Queries de verificación — ByteNova

Este notebook contiene las queries analíticas de verificación del modelo de datos de ByteNova. El objetivo es demostrar que el modelo funciona correctamente y que los datos tienen sentido de negocio.

## 1. Ingresos totales por mes

```python
query_ingresos_por_mes = """
SELECT 
    FORMAT_TIMESTAMP('%Y-%m', paid_at) AS mes,
    ROUND(SUM(amount), 2) AS ingresos_totales
FROM `bytenova_db.payments`
WHERE status = 'completed'
GROUP BY mes
ORDER BY mes
"""
```

## 2. Productos más vendidos

```python
query_productos_mas_vendidos = """
SELECT 
    p.name AS producto,
    SUM(oi.quantity) AS unidades_vendidas
FROM `bytenova_db.order_items` oi
JOIN `bytenova_db.products` p ON oi.product_id = p.product_id
GROUP BY producto
ORDER BY unidades_vendidas DESC
LIMIT 10
"""
```

## 3. Clientes por país

```python
query_clientes_por_pais = """
SELECT 
    country AS pais,
    COUNT(customer_id) AS total_clientes
FROM `bytenova_db.customers`
GROUP BY pais
ORDER BY total_clientes DESC
"""
```

## 4. Tiempo medio de entrega

```python
query_tiempo_medio_entrega = """
SELECT 
    ROUND(AVG(TIMESTAMP_DIFF(delivered_at, ordered_at, DAY)), 2) AS dias_entrega_promedio
FROM `bytenova_db.orders`
WHERE delivered_at IS NOT NULL
"""
```

## 5. Productos mejor valorados

```python
query_productos_mejor_valorados = """
SELECT 
    p.name AS producto,
    ROUND(AVG(r.rating), 2) AS valoracion_media,
    COUNT(r.review_id) AS total_valoraciones
FROM `bytenova_db.reviews` r
JOIN `bytenova_db.order_items` oi ON r.order_item_id = oi.order_item_id
JOIN `bytenova_db.products` p ON oi.product_id = p.product_id
GROUP BY producto
HAVING total_valoraciones >= 5
ORDER BY valoracion_media DESC
LIMIT 10
"""
```
